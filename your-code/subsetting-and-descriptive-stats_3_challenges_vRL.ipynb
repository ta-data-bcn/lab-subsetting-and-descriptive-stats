#### Group the employees by number of years of employment. What are the average salaries in each group?

```python
#Grouping by years to see, for each year, how many employees belong to each Year category

years_groupby = employees.groupby('Years').agg('count')

years_groupby
```

####  What is the average salary per title?

```python
#Grouping By Titles, & Years in the company, gives the average salary for each title
years_groupby = employees.groupby(['Title', 'Years']).agg({'Salary':'mean'})
years_groupby
```
